```python
#@title 기본 제목 텍스트
import pandas as pd
from scipy.stats import chi2_contingency
url= 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/datasets/vgames2.csv'
df= pd.read_csv(url)

df
```

```python
#df= df.iloc[0:,1:]
#df= df.sort_values(by='Year',ascending=False)
#df
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 0.001
        return 0.001
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1
        return 1
    return x

def year_(x) :
    if x < 20 :
        return x + 2000
    if    100  > x > 20 :
        return x + 1900
    else : 
        return x   # 연도표기가 잘못된 데이터 전처리
df['Year']= df['Year'].apply(year_)


df['NA_Sales'] = df['NA_Sales'].apply(value_to_float)
df['EU_Sales']= df['EU_Sales'].apply(value_to_float)
df['JP_Sales']= df['JP_Sales'].apply(value_to_float)
df['Other_Sales']= df['Other_Sales'].apply(value_to_float)                
df['NA_Sales'] = df['NA_Sales'].astype(float)
df['EU_Sales'] = df['EU_Sales'].astype(float)
df['JP_Sales'] = df['JP_Sales'].astype(float)
df['Other_Sales'] = df['Other_Sales'].astype(float)          #타입케스팅 
df.loc[:,'Whole_Sales']=df['Other_Sales']+ df['JP_Sales'] +df['EU_Sales'] + df['NA_Sales']   # 게임 전체판매량
df=df.iloc[0:,1:]
df['Genre Code']= pd.factorize(df['Genre'])[0] #장르별로 숫자를 매겨서 장르코드 만들기
#df.dropna(subset=['Year'])
#df.head(40)

df=df.dropna(axis=0, how='any')
df['Year']=df['Year'].astype(int)


df=df.reset_index()
df=df.iloc[0:,1:]
df
#데이터 전처리 끝
```
